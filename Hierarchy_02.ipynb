```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features: Sepal length, Sepal width, Petal length, Petal width
iris_df = pd.DataFrame(X, columns=iris.feature_names)

# Plot Dendrogram
plt.figure(figsize=(10, 6))
dend=sch.dendrogram(sch.linkage(iris_df, method='ward'))
plt.title("Dendrogram for Iris Dataset")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
# Print the color_list from the dendrogram
print(dend['color_list'])
unique_colors=set(dend['color_list'])
print(unique_colors)

totol_cluster=len(unique_colors)-1
print(totol_cluster)
# unique_colors=set(den)

# Selecting the Petal length and Petal width features
data = X[:, 2:4]

# Agglomerative Clustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
labels_ = cluster.fit_predict(data)

# Scatter plot for Agglomerative Clustering with Petal length and Petal width
plt.scatter(data[:, 0], data[:, 1], c=labels_, cmap='rainbow')
plt.title('Agglomerative Clustering of Iris Data (Petal Length and Petal Width)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features: Sepal length, Sepal width, Petal length, Petal width
y = iris.target  # True species labels (Setosa, Versicolor, Virginica)

# Perform Agglomerative Clustering with n_clusters=3 (since there are 3 species)
cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
labels_ = cluster.fit_predict(X)

# Compare clustering results with true species labels
print(f"True species labels:\n{y}")
print(f"Clustered labels:\n{labels_}")

# Visualize the clustering (Petal Length vs Petal Width for better separation)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 2], X[:, 3], c=labels_, cmap='rainbow', label='Clustered Data')
plt.title('Agglomerative Clustering of Iris Data (Petal Length vs Petal Width)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()

# Optional: Visualize the true species labels to compare
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 2], X[:, 3], c=y, cmap='rainbow', label='True Labels')
plt.title('True Species Labels of Iris Data (Petal Length vs Petal Width)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()

# Print the color_list from the dendrogram
print(dend['color_list'])
unique_colors = set(dend['color_list'])
print(unique_colors)

total_cluster = len(unique_colors) - 1
print(f"Total clusters from Dendrogram: {total_cluster}")
```
